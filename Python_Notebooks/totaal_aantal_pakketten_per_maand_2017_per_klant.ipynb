```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import data
df = pd.read_csv('/datc/parcel/notebooks/data/postnl/20200904_sample.csv', index_col=0, parse_dates=[1])
```

```python
# Manipulate data
df['maand'] = df['procesdag'].dt.month
new_df = df.where(df['procesdag'].dt.year == 2017).groupby([df['cust_id'], df['maand']]).agg(totaal_aantal_pakketten = ('aantal_pakketten', 'sum'), totaal_aantal_pakketten_volgende_dag = ('aantal_pakketten_volgende_dag', 'sum')).reset_index()
```

```python
# Total number of packages per month 2017
df_klant1 = new_df.where(new_df['cust_id'] == 'klant_1').dropna()
fig, ax = plt.subplots()
ax.bar(df_klant1['maand'], df_klant1['totaal_aantal_pakketten'])
ax.set(
    xlabel = 'Maand',
    ylabel = 'Aantal pakketten',
    title = 'Totaal aantal pakketten per maand 2017 voor klant 1')
```
