```python
# Import Packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import Data
data = pd.read_csv('/datc/parcel/notebooks/data/postnl/20201014_300_klanten.csv', parse_dates=[0])
```

```python
data
```

```python
# Manipulate Data
cust_filter = data['cust_id'] == 'klant_69'
date_filter = data['procesdag'].dt.dayofweek <= 4
data = data.where(date_filter & cust_filter).dropna()
```

```python
data
```

```python
y = data.aantal_pakketten_volgende_dag
X = data[['procesdag', 'aantal_pakketten']]
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(data['procesdag'], data['aantal_pakketten'])
plt.xlabel = ('Procesdag')
plt.ylabel = ('Aantal Pakketten')
plt.title = ('Aantal Pakketten Klant 69')
plt.grid()
plt.show()
```
